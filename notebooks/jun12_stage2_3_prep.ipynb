### Literature Review — Stage 2 (Synchronization & ETF Microstructure)

**Kirilenko, Kraus, Linton & Xiao (2025) — ETF (Mis)pricing**
Cambridge Working Papers in Economics. Using regulatory data covering 128 ETFs
from 2018–2022, this paper shows that ETF mispricing (price vs NAV gap) is directly
linked to Authorized Participants' (APs) inventory constraints. When APs approach
their inventory limits — especially during market stress — their ability to arbitrage
away mispricing breaks down, causing structural price distortions. Directly supports
our argument that PBP's end-of-day TE is not just a timing issue but is amplified by
AP inventory dynamics near market close.

**Koont, Ma, Pastor & Zeng (2022) — Steering a Ship in Illiquid Waters**
NBER Working Paper. Shows that passive ETFs actively deviate from strict index
tracking to manage liquidity — they adjust creation/redemption baskets to absorb
shocks from investors. Tracking error is therefore partly a deliberate feature of
liquidity management, not just poor replication. Supports our finding from the PBP
Prospectus that smearing and sampling are intentional, not accidental.

### Literature Review — Stage 3 (Volatility Risk Premium)

**Bakshi & Kapadia (2003) — Delta-Hedged Gains and the Negative Market VRP**
Review of Financial Studies. Seminal paper proving that IV systematically exceeds RV
for S&P 500 options. Using delta-hedged portfolios, they show that option buyers
consistently pay a premium to hedge against volatility spikes — meaning the volatility
risk premium (VRP) is negative. At-the-money calls lose on average ~0.10% of index
level per hedging period. This is the core theoretical justification for why selling ATM
calls (BXM strategy) is structurally profitable — you are the one collecting this premium.

**Broadie, Chernov & Johannes (2007) — Model Specification and Risk Premia**
Journal of Finance. Using S&P 500 futures options, shows that option prices embed
both diffusive volatility risk premia AND jump risk premia. Investors demand extra
compensation for sudden price jumps and volatility spikes — this shapes the implied
volatility smile. Relevant to Stage 3 because the premium collected by BXM when
selling ATM calls is not just compensation for general variance, but also for bearing
jump risk.

**Ahmad & Wilmott (2008) — Which Free Lunch Would You Like Today, Sir?**
Wilmott Magazine. Shows that hedging with implied volatility (IV) vs realized
volatility (RV) produces fundamentally different risk/return profiles. When IV > RV
(options overpriced), selling options and delta-hedging with IV generates a path-
dependent but on-average positive profit. Directly motivates our Stage 3 dynamic
switching algorithm — when IV-RV spread is large, lean into the BXM strategy.
