### Literature Review (June 2nd ~ June 12nd)

**Kirilenko, Kraus, Linton & Xiao (2025) — ETF (Mis)pricing**
*Cambridge Working Papers in Economics*
Uses regulatory data on 128 ETFs (2018–2022) to show that ETF mispricing (price vs NAV)
is directly linked to Authorized Participants' (APs) inventory constraints. When APs hit
their inventory limits during market stress, their ability to arbitrage away mispricing breaks
down — causing structural price distortions unrelated to fundamentals. Most relevant to us:
this mechanism is most active near market close, which directly supports why our 15-minute
synchronization window matters. The COVID-19 period showed bond ETF mispricings
exceeding 5%, prompting Fed intervention via the SMCCF.

**Koont, Ma, Pastor & Zeng (2022) — Steering a Ship in Illiquid Waters**
*NBER Working Paper*
Shows that passive ETFs intentionally deviate from strict index tracking to manage
liquidity — they adjust creation/redemption baskets to absorb shocks from investors.
Tracking error is therefore partly a deliberate feature of liquidity management, not just
poor replication. Key finding: illiquidity of the underlying index is positively and
significantly associated with higher tracking error (β significant at 1% level across all
specs). Directly supports our Prospectus finding that PBP's smearing and sampling are
intentional structural features, not accidents.

**Bakshi & Kapadia (2003) — Delta-Hedged Gains and the Negative Market VRP**
*Review of Financial Studies*
Seminal paper proving that IV systematically exceeds RV for S&P 500 options. Using
delta-hedged call portfolios (S&P 500, 1988–1995), they show that option buyers
consistently pay a premium to hedge against volatility — the volatility risk premium (VRP)
is negative. ATM delta-hedged calls lose on average ~$0.43 per contract (~0.10% of index
level). The losses deepen at higher volatility regimes and longer hedging horizons.
Core theoretical justification for why selling ATM calls (BXM) is structurally profitable —
the option seller is collecting this VRP.

**Broadie, Chernov & Johannes (2007) — Model Specification and Risk Premia**
*Journal of Finance*
Using S&P 500 futures options data, identifies that option prices embed both diffusive
volatility risk premia AND jump risk premia. Strong evidence that investors demand
compensation for sudden price jumps and volatility spikes — which shapes the implied
volatility smile. The SVJ and SVCJ models significantly outperform the pure SV model,
confirming jump risk is a meaningful priced factor. Relevant to Stage 3: the premium
collected by BXM selling ATM calls is not just compensation for general variance, but
also for bearing jump risk. This adds nuance to the IV-RV spread interpretation.

**Kumiega, Sterijevski & Wills (2024) — Black-Scholes 50 Years Later**
*International Journal of Financial Studies*
Re-evaluates passive option strategies (BXM, BXMD, BXMC, PUT, etc.) over 2012–2023.
Key finding: simple BXM underperforms S&P 500 on both absolute and risk-adjusted basis
in this period (Annual Return 5.97%, Return-to-Risk 0.4629 vs SPX 0.6128). However,
nuanced strategies with conditional logic (BXMC — switches between BXM and BXMH
based on VIX > 20) and BXMH outperform. Directly motivates our Stage 3 dynamic
switching algorithm: static BXM is not enough — adding IV-based logic improves
risk-adjusted returns. Also confirms BXM is most effective during high-volatility regimes.
